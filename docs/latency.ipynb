```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
x = np.linspace(0, 5, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title(r'$y=\sin(x)$')  # Use LaTeX for the title
plt.xlabel(r'$x$')  # Use LaTeX for the x-label
plt.ylabel(r'$y$')  # Use LaTeX for the y-label

plt.show()
```
